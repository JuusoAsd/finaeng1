#### 1. Compute the difference A between the change in the value of the option and the change in the value the replicating portfolio

```python
def d_1(S, E, r, T, Cobs):
    vola = .5
    d = 1
    eps = np.finfo(float).eps
    
    while(d > eps):
        d1 = (np.log(S / E) + (r + vola ** 2 / 2) * T) / (vola * np.sqrt(T) + eps)
        d2 = d1 - vola * np.sqrt(T)
        C = S * norm.cdf(d1) - E * np.exp(-r * T) * norm.cdf(d2)
        vega = S * np.sqrt(T) * np.exp(-d1 ** 2 / 2) / (np.sqrt(2 * np.pi) + eps)
        d = (Cobs - C) / -vega
        vola = vola - d
    return d1

    
# Define find_delta() method to calculate the delta values  

def find_delta(S, E, Cobs, r, T):  
    
    # Estimate the d1 by computing the implied volatility
    d1 = d_1(S, E, r, T, Cobs)

    # Use the estimated d1 to compute the delta
    delta = norm.cdf(d1)
    return delta
```

```python
# Compute the delta for each row

# Initialize df_2 to include df_1 and delta values 

df_2 = df_1.reset_index()
df_2["Delta"] =""

for index, row in df_2.iterrows():
    delta = find_delta(row["S"],row["E"],row["Cobs"], row["r"], row["T"])
    df_2["Delta"].iloc[index] = delta

plt.plot(df_2['Delta'])
print(df_2.to_markdown())
```
